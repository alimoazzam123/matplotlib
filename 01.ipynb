```python
batsman=pd.read_csv('sharma-kohli.csv')
batsman
```

```python
#simple plot
plt.plot(batsman['index'],batsman['V Kohli'])
```

```python
#multiple plots
plt.plot(batsman['index'],batsman['V Kohli'])
plt.plot(batsman['index'],batsman['RG Sharma'])
```

```python
#color(hex) and line(width and style) and marker(size)

plt.plot(batsman['index'],batsman['V Kohli'],color='green')
plt.plot(batsman['index'],batsman['RG Sharma'],color='black')

plt.title('Rohit sharma virat kohli carrer comparision')
plt.xlabel('Season')
plt.ylabel('Runs Score')
```

```python
plt.plot(batsman['index'],batsman['V Kohli'],color='green',linestyle='dashed',linewidth=4,marker='D',markersize=10)
plt.plot(batsman['index'],batsman['RG Sharma'],color='black',linestyle='dashed',linewidth=2,marker='>',markersize=5)

plt.title('Rohit sharma virat kohli carrer comparision')
plt.xlabel('Season')
plt.ylabel('Runs Score')
```

```python
#legend->loaction
plt.plot(batsman['index'],batsman['V Kohli'],color='green',linestyle='dashed',linewidth=4,marker='D',markersize=10,label='Virat')
plt.plot(batsman['index'],batsman['RG Sharma'],color='black',linestyle='dashed',linewidth=2,marker='>',markersize=5,label='Rohit')

plt.title('Rohit sharma virat kohli carrer comparision')
plt.xlabel('Season')
plt.ylabel('Runs Score')
plt.legend(loc='best')
```

```python
#limiting axes
price=[48000,54000,57000,49000,47000,45000,450000]
year=[2015,2016,2017,2018,2019,2020,2021]

plt.plot(year,price)
plt.ylim(0,100000)
```

```python
#grid
plt.plot(batsman['index'],batsman['V Kohli'],color='green',linestyle='dashed',linewidth=4,marker='D',markersize=10,label='Virat')
plt.plot(batsman['index'],batsman['RG Sharma'],color='black',linestyle='dashed',linewidth=2,marker='>',markersize=5,label='Rohit')

plt.title('Rohit sharma virat kohli carrer comparision')
plt.xlabel('Season')
plt.ylabel('Runs Score')
plt.legend(loc='best')
plt.grid()
```

Piechart

Univariate/Bivariate Analysis
Categorical vs numerical
Use case - To find contibution on a standard scale

```python
# simple data
data = [23,45,100,20,49]
subjects = ['eng','science','maths','sst','hindi']
plt.pie(data,labels=subjects)

plt.show()
```

```python
# percentage and color
data = [23,45,100,20,49]
subjects = ['eng','science','maths','sst','hindi']
plt.pie(data,labels=subjects,autopct='%0.1f%%',colors=['blue','green','yellow','pink','cyan'])

plt.show()
```

```python
#explode shadow
data = [23,45,100,20,49]
subjects = ['eng','science','maths','sst','hindi']
plt.pie(data,labels=subjects,autopct='%0.1f%%',colors=['blue','green','yellow','pink','cyan'],explode=[0.1,0,0,0,0])

plt.show()
```
